```python
import numpy as np
import matplotlib.pyplot as plt

N = 100
ux1 = 2
vx1 = 3
uy1 = 1
vy1 = 1

x1 = np.random.normal(ux1, vx1, N)
y1 = np.random.normal(uy1, vy1, N)


ux2 = 8
vx2 = 3
uy2 = 6
vy2 = 1

x2 = np.random.normal(ux2, vx2, N)
y2 = np.random.normal(uy2, vy2, N)


plt.axis('equal')
plt.scatter(x1, y1)
plt.scatter(x2, y2)


# mu1= 0.5
# mu2= 1
# var1= 0.8
# var2= 0.6
# np.random.normal

# #muestras

# x1= np.random.normal(mu1, var1, N1)
# x2= np.random.normal(mu2, var2, N2)

# muy1=6
# vary1=0.7

# y1=np.random.normal(muy1, vary1, N1)
# plt.scatter(x1,y1)

# muy2=0.2
# vary2=0.8
# #Para correr varianza más chiquita distancia eje vertical

# #También se puede utilizar media distancia eje horizantal

# y2=np.random.normal(muy2, vary2, N2)
# plt.scatter(x2, y2)
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Crear subgráficos en una matriz 2x2
plt.subplot(2, 2, 1)
plt.axis('equal')
plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.title('')

plt.subplot(2, 2, 2)
plt.hist(x,bins=2)
plt.title('')

plt.subplot(2, 2, 3)
plt.hist(y,bins=30)
plt.title('')

plt.subplot(2, 2, 4)
plt.hist(x1, bins=30)
plt.title('')

# Ajustar el diseño para evitar superposiciones
plt.tight_layout()

# Mostrar los subgráficos
plt.show()
```
